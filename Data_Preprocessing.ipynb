```python
import pandas as pd
from collections import Counter
import random
import csv
```

```python
data = pd.read_excel("AHD.xlsx")
data
```

```python
print(data.head())
print(data.info())
```

```python
# Drop duplicates
data = data.drop_duplicates()
print(data.info())
```

```python
category_counts = data["Category"].value_counts()
print("Original Category Distribution:")
print(category_counts.to_string())
```

```python
# Parameters
total_samples = 3000
min_samples_per_category = 1  # Ensures at least one sample is taken if the category exists
```

```python
# Calculate the number of samples per category
category_ratios = category_counts / category_counts.sum()
samples_per_category = (category_ratios * total_samples).round().astype(int)
print(category_ratios)
print(samples_per_category)
```

```python
# Ensure a minimum number of samples per category
samples_per_category = samples_per_category.clip(lower=min_samples_per_category)

# Stratified sampling
sampled_data = pd.DataFrame()
for category, n_samples in samples_per_category.items():
    category_data = data[data["Category"] == category]
    sampled_category_data = category_data.sample(
        n=min(n_samples, len(category_data)), random_state=42
    )
    sampled_data = pd.concat([sampled_data, sampled_category_data])

# Shuffle the sampled data to mix categories
sampled_data = sampled_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Inspect the new distribution
print("Sampled Category Distribution:")
print(sampled_data["Category"].value_counts())

# Save as a CSV file
sampled_data.to_csv("sampled_arabic_qa.csv")
print("Balanced sampled data saved successfully as CSV!")
```

```python
from arabert.preprocess import ArabertPreprocessor

# Initialize AraBERT preprocessor
model_name = "aubmindlab/bert-base-arabertv2"
arabert_prep = ArabertPreprocessor(model_name=model_name)
```

```python
# Define preprocessing function
def preprocess_arabic_arabert(text):
    if pd.isnull(text):  # Handle missing values
        return ""
    return arabert_prep.preprocess(text)

# Apply preprocessing to Question and Answer columns
df["Question"] = df["Question"].apply(preprocess_arabic_arabert)
df["Answer"] = df["Answer"].apply(preprocess_arabic_arabert)

# Save preprocessed data to a new CSV
output_file = "preprocessed_arabic_qa.csv"
df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"Preprocessing complete! Data saved to '{output_file}'.")
```
